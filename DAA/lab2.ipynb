```python
import matplotlib.pyplot as plt
import numpy as np

# (a) logn
def logn(n):
    return np.log(n)

# (b) log2 n
def log2n(n):
    return np.log2(n)

# (c) 3n
def three_n(n):
    return 3*n

# (d) nlogn
def nlogn(n):
    return n*np.log(n)

# (e) n 2
def n_2(n):
    return n**2

# (f) 2n
def two_n(n):
    return 2**n

# (g) n!
def n_fact(n):
    res = np.zeros_like(n)
    for i in range(len(n)):
        res[i] = np.math.factorial(n[i])
    return res

n = np.arange(1, 10, 1)
plt.plot(n, logn(n), label='logn')
plt.plot(n, log2n(n), label='log2n')
plt.plot(n, three_n(n), label='3n')
plt.plot(n, nlogn(n), label='nlogn')
plt.plot(n, n_2(n), label='n^2')
plt.plot(n, two_n(n), label='2^n')
plt.plot(n, n_fact(n), label='n!')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.legend()
plt.show()
```

```python
def check_distinct(lst):
    n = len(lst)
    for i in range(n):
        for j in range(i+1, n):
            if lst[i] == lst[j]:
                return False
    return True

list1 = [1,2,3,4,5,6,7,7]
list2 = [1,2,3,4,5]
print("For List 1: ", check_distinct(list1))
print("For List 2: ", check_distinct(list2))
```

```python
def majorityElement(nums):
    freq = {}
    max = 0
    maxCount = 0
    for num in nums:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
        if freq[num]>maxCount:
            max = num
            maxCount = freq[num]
    return max
        
        
arr = list(map(int, input().split(' ')))
print(arr)
print('The majority element is : ', majorityElement(arr))
```
