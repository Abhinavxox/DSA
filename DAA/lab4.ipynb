```python
def currency_exchange(amount, denominations):
    denominations.sort(reverse=True)
    result = []
    for denomination in denominations:
        while denomination <= amount:
            result.append(denomination)
            amount -= denomination
    if amount > 0:
        return "No optimal solution found."
    return result

test_cases = [
    (73, [1, 2, 5, 10, 20, 50]),
    (123, [1, 2, 5, 10, 20, 50]),
    (78, [1, 2, 5, 10, 20, 50]),
    (84, [1, 5, 10, 20, 50]),
    (30,[7])
]

for test in test_cases:
    amount, denominations = test
    result = currency_exchange(amount, denominations)
    print(f"Amount: {amount}, Denominations: {denominations}")
    print(f"Optimal Solution: {result}")
    print()
```

```python
def schedule_tasks(tasks):
    tasks.sort(key=lambda x: x[1])
    schedule = [tasks[0]]

    for task in tasks[1:]:
        if task[0] >= schedule[-1][1]:
            schedule.append(task)

    return schedule

tasks = [
    (4, 9),
    (2, 5),
    (1, 7),
    (6, 8),
    (3, 10)
]

scheduled_tasks = schedule_tasks(tasks)
for task in scheduled_tasks:
    print(f"Task: {task}")
```

```python
from typing import List

class Solution:
    def canJump(self, nums: List[int]) -> bool:
        last_position = len(nums) - 1

        for i in range(len(nums) - 2, -1, -1):
            if (i + nums[i]) >= last_position:
                last_position = i
        return last_position == 0

nums = [3,2,1,0,4]
s = Solution()
print(s.canJump(nums))
```

```python
class Solution:
    def findContentChildren(self, g: List[int], s: List[int]) -> int:
        g.sort()
        s.sort()
        i, j = 0, 0
        while i < len(g) and j < len(s):
            if s[j] >= g[i]:
                i += 1
            j += 1
        return i

g = [1,2,3]
s = [1,1]
soln = Solution()
print(soln.findContentChildren(g,s))
```

```python
from re import S
class Solution:
    def numRescueBoats(self, people: List[int], limit: int) -> int:
        people.sort()
        lo = 0
        hi = len(people)-1
        boats = 0
        while lo <= hi:
            if people[lo] + people[hi] <= limit:
                lo += 1
                hi -= 1
            else:
                hi -= 1
            boats += 1
        return boats

people = [3,2,2,1]
limit = 3

s = Solution()
print(s.numRescueBoats(people,limit))
```
